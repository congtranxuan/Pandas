# HEROES OF PYMOLI DATA ANALYSIS

1. Most players are male compared with female and other.
2. Almost half of all players is in range of 20 to 24.
3. The best-sell product is Oathbreaker, Last Hope of the Breaking Storm and it is most profitable as well.

```python
#Dependancies and Set up
import pandas as pd

#Import the data file to the 
filepath = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(filepath)
purchase_data.head()
```

   # Player Count
       Total Number of Players

```python
#Name_player = purchase_data["SN"].value_counts()
Number_player = purchase_data["SN"].value_counts().count()
outcome_playercount = pd.DataFrame({"Total Players": [Number_player]})
outcome_playercount
```

# Purchasing Analysis (Total)
    Number of Unique Items
    Average Purchase Price
    Total Number of Purchases
    Total Revenue

```python
Items_number = purchase_data["Item ID"].value_counts().count()
Average_price = purchase_data["Price"].mean()
Purchase_number = purchase_data["Purchase ID"].count()
Total_revenue = "${:.2f}".format(Average_price*Purchase_number)
Purchasing_analysis = pd.DataFrame({"Number of Unique Items":[Items_number],"Average Price":["${:.2f}".format(Average_price)],"Number of Purchases":[Purchase_number],"Total Revenue":[Total_revenue]})
Purchasing_analysis
```

# Gender Demographics
    Percentage and Count of Male Players
    Percentage and Count of Female Players
    Percentage and Count of Other / Non-Disclosed

```python
Gender = purchase_data[["SN","Gender"]].groupby("SN").head(1)["Gender"].value_counts()
Number_player = purchase_data["SN"].value_counts().count()
Outcome_Genderdemo = pd.DataFrame({"":["Male","Female","Other/ Non-Disclosed"],"Total Count":[Gender[0],Gender[1],Gender[2]],"Percentage of Players":["{:.2f}%".format(Gender[0]*100/Number_player), "{:.2f}%".format(Gender[1]*100/Number_player), "{:.2f}%".format(Gender[2]*100/Number_player)]})
Outcome_Genderdemo
```

# Purchasing Analysis (Gender)
    The below each broken by gender
    Purchase Count
    Average Purchase Price
    Total Purchase Value
    Average Purchase Total per Person by Gender

```python
Gender_grouped = purchase_data.groupby(["Gender"])
Purchase_count = Gender_grouped.count()
Ave_price = Gender_grouped.mean()
Total_purchase = Gender_grouped.sum()
No_player_gender = Gender_grouped.nunique()
Outcome_gender_sale = pd.DataFrame({" ":["Gender","Female","Male","Other/ Non-Disclosed"],
                                    "Purchase Count":["",Purchase_count["Purchase ID"][0],Purchase_count["Purchase ID"][1],Purchase_count["Purchase ID"][2]],
                                    "Average Purchase Price":["","${:.2f}".format(Ave_price["Price"][0]),"${:.2f}".format(Ave_price["Price"][1]),"${:.2f}".format(Ave_price["Price"][2])],
                                    "Total Purchase Value":["","${:.2f}".format(Total_purchase["Price"][0]),"${:.2f}".format(Total_purchase["Price"][1]),"${:.2f}".format(Total_purchase["Price"][2])],
                                    "Avg Total Purchase per Person":["","${:.2f}".format(Total_purchase["Price"][0]/No_player_gender["SN"][0]),"${:.2f}".format(Total_purchase["Price"][1]/No_player_gender["SN"][1]),"${:.2f}".format(Total_purchase["Price"][2]/No_player_gender["SN"][2])]})
Outcome_gender_sale
```

## Age Demographics
    The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
    Purchase Count
    Average Purchase Price
    Total Purchase Value
    Average Purchase Total per Person by Age Group

```python
purchase_data.head()
bins = [0,9,14,19,24,29,34,39,100]
label = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Range of Age"] = pd.cut(purchase_data["Age"],bins, labels = label)
Groupbyage=purchase_data.groupby(["Range of Age"])
Age_demo = Groupbyage.nunique()
Number_player = purchase_data["SN"].value_counts().count()
Outcome_age = pd.DataFrame({"Total Count":Age_demo["SN"],"Percentage of Players":round(Age_demo["SN"]*100/Number_player,2)})
Outcome_age
```

```python
Outcome_sale_age = pd.DataFrame({"Purchase Count":Groupbyage.count()["Purchase ID"],"Average Purchase Price":Groupbyage.mean()["Price"].map("${:.2f}".format),"Total Purchase Value":(Groupbyage.sum()["Price"]).map("${:.2f}".format),"Avg Total Purchase per Person":round(Groupbyage.sum()["Price"]/Groupbyage.nunique()["SN"],2).map("${:.2f}".format)})
Outcome_sale_age
```

## Top Spenders
    Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
    SN
    Purchase Count
    Average Purchase Price
    Total Purchase Value

```python
GroupbySN = purchase_data.groupby("SN")
Outcome_spender = pd.DataFrame({"Purchase Count":GroupbySN.count()["Purchase ID"],
                                "Average Purchase Price":round(GroupbySN.mean()["Price"],2).map("${:.2f}".format),
                                "Total Purchase Value":(GroupbySN.mean()["Price"]*GroupbySN.count()["Purchase ID"]).astype(float)}).sort_values("Total Purchase Value", ascending = False)
Outcome_spender["Total Purchase Value"] = Outcome_spender["Total Purchase Value"].map("${:.2f}".format)
Outcome_spender.head(5)
```

## Most Popular Items
    Identify the 5 most popular items by purchase count, then list (in a table):
    Item ID
    Item Name
    Purchase Count
    Item Price
    Total Purchase Value

```python
Groupbyitem = purchase_data.groupby(["Item ID","Item Name"])
Outcome_item = pd.DataFrame({"Purchase Count":Groupbyitem.count()["Purchase ID"],
                             "Item Price":Groupbyitem.mean()['Price'].map("${:.2f}".format),
                             "Total Purchase Value":(Groupbyitem.count()["Purchase ID"]*Groupbyitem.mean()['Price']).astype(float)}).sort_values("Purchase Count",ascending = False)
Outcome_item["Total Purchase Value"] = Outcome_item["Total Purchase Value"].map("${:.2f}".format)
Outcome_item.head()
```

## Most Profitable Items
    Identify the 5 most profitable items by total purchase value, then list (in a table):
    Item ID
    Item Name
    Purchase Count
    Item Price
    Total Purchase Value

```python
Outcome_item = pd.DataFrame({"Purchase Count":Groupbyitem.count()["Purchase ID"],
                             "Item Price":Groupbyitem.mean()['Price'].map("${:.2f}".format),
                             "Total Purchase Value":(Groupbyitem.count()["Purchase ID"]*Groupbyitem.mean()['Price']).astype(float)}).sort_values("Purchase Count",ascending = False)
Outcome_item["Total Purchase Value"]=Outcome_item["Total Purchase Value"].astype(float)
Outcome_item = Outcome_item.sort_values("Total Purchase Value",ascending = False)
Outcome_item["Total Purchase Value"]=Outcome_item["Total Purchase Value"].map("${:.2f}".format)
Outcome_item.head()
```
